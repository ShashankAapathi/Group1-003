```python
df= df.drop(columns = "Unnamed: 0")
```

```python
df["Salary Estimate"] = df["Salary Estimate"].replace("(Glassdoor est.)", "",regex=True)
```

```python
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: x.strip('()'))
```

```python
df["Size"] = df["Size"].replace("(employees)", "",regex=True)
```

```python
df
```

```python
plt.figure(figsize = (15,5))

df["Company Name"].value_counts().sort_values(ascending=False).head(20).plot.bar(color = "orange",fontsize=12, edgecolor='Red', linewidth = 2)
plt.title("Most popular companies by Data Scientist Jobs")
plt.xlabel("Company Name",fontsize=15)
plt.ylabel("Count")

plt.show()
```

```python
sns.countplot(x='Rating', data = df)
plt.show()
```

```python
plt.figure(figsize = (10,5))
df["Sector"].value_counts().sort_values(ascending=False).plot.bar(color= "red",fontsize=12)
plt.title("Different types of Sectors in DataScience")
plt.xlabel("Sectors",fontsize=15)
plt.ylabel("Count")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,5)
color = plt.cm.BuPu(np.linspace(0,1,20))
df["Job Title"].value_counts().sort_values(ascending=False).head(20).plot.bar(color=color)
plt.title("Job Title Offered",fontsize=20)
plt.xlabel("Job Title",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
chart = sns.barplot(
    data=df,
    x='Revenue',
    y='Rating'
)
chart.set_xticklabels(
    chart.get_xticklabels(), 
    rotation=65, 
    horizontalalignment='right',
    fontweight='light',
 
)
chart.axes.yaxis.label.set_text("Companies Rating")
```

```python
corr = df.corr()

ax = sns.heatmap(
corr,
vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(20, 220, n=200),
square=True
)
ax.set_xticklabels(
ax.get_xticklabels(),
rotation=45,
horizontalalignment='right'
);
ax.set_title("Plot for displaying correlation between various columns",color ="Darkred" ,fontsize = 22)
plt.xlabel("Column Names",fontsize=17,color = "GREEN")
```

```python
plt.rcParams['figure.figsize'] = (12,9)
color = plt.cm.PuRd(np.linspace(0,1,20))
df["Founded"].value_counts().sort_values(ascending=False)[1:21].plot.bar(color=color)
plt.title("Number of Companies established in the single Year",fontsize=20)
plt.xlabel("Foundation Year",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (12,9)
color = plt.cm.BuPu(np.linspace(0,1,20))
df["Location"].value_counts().sort_values(ascending=False).head(20).plot.bar(color=color)
plt.title("Top 20 locations for Data Science Job",fontsize=20)
plt.xlabel("Locations",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (12,9)
color = plt.cm.hsv(np.linspace(0,1,20))
df["Sector"].value_counts().sort_values(ascending=False).plot.bar(color=color)
plt.title("Different types of Sectors in DataScience",fontsize=20)
plt.xlabel("Sectors",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```
